```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(-10, 11)
b = np.arange(-10, 11)
c = np.arange(-10, 11)
a, b, c
```

# Ciclo *for*

```python
x = np.linspace(-25, 25, 100)
for i in a:
    for j in b:
        for k in c:
            y = i*x**2+j*x+k
            plt.plot(x,y)
```

# Meshgrid method

```python
x = np.linspace(-25, 25, 100)
A, B, C = np.meshgrid(a, b, c, indexing='xy')
x = x[:, np.newaxis, np.newaxis, np.newaxis]
y = A * x ** 2 + B * x + C
y = y.reshape(100, -1)
plt.plot(x[:, 0, 0, 0], y)
plt.show()
```
