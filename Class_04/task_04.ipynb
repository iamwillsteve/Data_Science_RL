```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('world-happiness-report.csv')
df.head()
```

```python
df._get_numeric_data().corr()
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df._get_numeric_data().corr('pearson'), annot=True, cmap='YlGnBu')
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 15))
sns.regplot(data = df, x='Overall rank', y='GDP per capita', ax=axs[0])
sns.regplot(data=df, x='Overall rank',
            y='Healthy life expectancy', ax=axs[1], color='green')
sns.regplot(data = df, x='Overall rank', y='Social support', ax=axs[2], color='orange')
plt.show()
```

Se observa una alta correlación de varias propiedades con la clasificación general. Las más destacadas son el PIB per cápita, la expectativa de vida saludable y el apoyo social.

Factores Económicos: El PIB per cápita es un factor significativo, indicando que un mayor rendimiento económico está asociado con una mayor felicidad.

Salud y Bienestar: La expectativa de vida saludable muestra una fuerte correlación, subrayando la importancia de la salud en la determinación de la felicidad.

Factores Sociales: El apoyo social y la libertad para tomar decisiones de vida son cruciales, reflejando cómo las estructuras sociales y la libertad personal contribuyen a la felicidad.
